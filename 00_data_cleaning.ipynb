```python
#Nuestro estudio será sobre los tiempos de las carreras de 100m, por lo que sobreescribimos 
#la tabla filtrando la columna "Event" con las carreras de 100m solamente.
#Desechamos las carreras de 200m, 400m, 800m, marathon, etc.

df1 = df1[df1.Event == '100 m']
df1 
```

```python
df1.sort_values (by = ['Time'])
```

```python
#nos fijamos que algunos tiempos no tienen 
#decimales ya que son exactos, por lo que procedemos a añadir decimales 
#para que todos ellos sigan el mismo patrón
df1[df1.Time == '00:00:10'] 
```
