```python
from dotenv import load_dotenv
import os
import requests
import re
import pandas as pd
import time
```

```python
df= pd.read_csv("./best_times_100m.csv", encoding='UTF-8')
df #insertamos el dataset con el que vamos a trabajar.
```

```python
load_dotenv()
def temperature(city, date):
    weather_token = os.getenv("weather_token")
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/weatherdata/history?aggregateHours=24&combinationMethod=aggregate&startDateTime={date}T00%3A00%3A00&endDateTime={date}T00%3A00%3A00&maxStations=-1&maxDistance=-1&contentType=json&unitGroup=metric&locationMode=single&key={weather_token}&dataElements=default&locations={city}"
    data = requests.get(url)
    weather = data.json()
    if "location" in weather:
        return weather["location"]["values"][0]["temp"]
    else:
        return None
```

```python
#por restricciones de la API en la capa gratuita, reducimos a 50 las filas a procesar
df = df.head(50)
df
```

```python
#iterrows() Iterate over DataFrame rows as (index, Series) pairs.
temps = []

for index, values in df.iterrows():
    city = values["City"]
    date = values["Date"]
    temp = temperature(city, date)
    print("Temperature for ", city, " at ", date, " was ", temp)
    temps.append(temp)
    time.sleep(1)
    
```

```python
temps
```

```python
df.insert(12, "Temp", temps)
df
```

```python
df.to_csv("./best_50_times_100m_with_city_temperature.csv")
```
